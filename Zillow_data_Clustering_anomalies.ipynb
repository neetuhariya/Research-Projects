```python
# Code snippet 2
# Assigning the data
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Zillow%20Tarrant%20County%20Raw%20Data.csv')

data
```

```python
# Code snippet 3
# Importing anomaly detection libraries
from pycaret.anomaly import *

```

```python
# Code snippet 4
# Setting up the pycaret config
anomaly_setup = setup(data)
```

```python
# Code snippet 5
# Building an anomaly detection model using knn algo
knn_model = create_model('knn')
```

```python
# Code snippet 6
# Identifying the anomalies using the knn model
knn_anamolies = predict_model(knn_model, data)
knn_anamolies
```

```python
# Code snippet 8
# Exporting anomalies to a csv file
knn_anamolies.to_csv("Zillow property anomalies.csv")
```

```python
# Code snippet 8.1
# Displaying anomalies
knn_anamolies[knn_anamolies['Anomaly'] == 1]
```

```python
# Code snippet 9
# Creating a new variable to store non-anomalies
clean_no_anomalies = knn_anamolies[knn_anamolies['Anomaly'] == 0]
clean_no_anomalies.shape
```

```python
# Code snippet 10
# Importing the clustering modules/algos
from pycaret.clustering import *
```

```python
# Code snippet 11
# Setting up the clustering configuration
cluster_setup = setup(clean_no_anomalies)
```

```python
# Code snippet 12
# Building a clustering ML model using kmeans algo
kmeans_model = create_model('kmeans')
```

```python
# Code snippet 13
# Identifying/assigning the clusters using the kmeans algo
kmeans_data = assign_model(kmeans_model)
```

```python
# Code snippet 14
# Displaying the data with clusters
kmeans_data
```

```python
# Code snippet 14
# Exporting the data with clusters to a csv file
kmeans_data.to_csv("Zillow data clusters.csv")
```

```python
# Code snippet 15
# Displaying the number of clusters
kmeans_data['Cluster'].value_counts()
```
