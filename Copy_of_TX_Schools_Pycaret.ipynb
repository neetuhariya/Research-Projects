```python
# Code snippet 3
# Setting the dependent variable/target
y =  'AveMonthSpend'
```

```python
# Code snippet 4
# Assigning noise/redundant/feeder/cheater variable to ignored variables list
ignored_cols  = ['CustomerID',  'Title', 'FirstName','BirthDate',
       'MiddleName', 'LastName', 'Suffix', 'AddressLine1', 'AddressLine2',

       'PhoneNumber']
```

```python
# Code snippet 5
# Assigning numerical features to a variable
num_cols  = ['NumberCarsOwned',
       'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
```

```python
# Code snippet 6
# Assigning categorical features to a variable
cat_cols = [ 'BikeBuyer',
       'City', 'StateProvinceName', 'CountryRegionName', 'PostalCode',
        'Education', 'Occupation', 'Gender',
       'MaritalStatus', 'HomeOwnerFlag']
```

```python
# Code snippet 7
# Importing all regression libraries form pycaret
from pycaret.regression import *
```

```python
# Code snippet 8
# Configuring/setting up pycaret with the data for ML modeling
regression_setup = setup(data,
                          target = y,
                         ignore_features = ignored_cols ,
                         numeric_features= num_cols,
                         categorical_features = cat_cols)
```

```python
# Code snippet 9
# Invoking ML algorithms
compare_models()
```
